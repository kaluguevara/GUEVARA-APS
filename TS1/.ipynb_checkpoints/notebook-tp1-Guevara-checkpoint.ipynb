#### **Ecuación 1: Identidad trigonométrica utilizada**
La siguiente identidad relaciona productos de senos con sumas de cosenos:

$$
2 \cdot \sin(\alpha)\sin(\beta) = \cos(\alpha-\beta) - \cos(\alpha+\beta)
$$

#### **Ecuación 2: Ortogonalidad de señales**
Dos señales discretas \( x[n] \) y \( y[n] \) son ortogonales si:

$$
\langle x, y \rangle = \sum_n x[n] \cdot y[n] \approx 0
$$

#### **Ecuación 3: Autocorrelación**
La autocorrelación de una señal \( x[n] \) se define como:

$$
R_{xx}[k] = \sum_{n=-\infty}^{\infty} x[n] \, x[n+k]
$$

#### **Ecuación 4: Correlación cruzada**
La correlación cruzada entre dos señales \( x[n] \) y \( y[n] \) está dada por:

$$
R_{xy}[k] = \sum_{n=-\infty}^{\infty} x[n] \, y[n+k]
$$

#### **Ecuación 5: Energía total de una señal discreta**
Para una señal discreta \( x[n] \) de duración finita o no periódica:

$$
E = \sum_{n=-\infty}^{\infty} |x[n]|^2
$$

#### **Ecuación 6: Potencia promedio de una señal periódica**
Para una señal discreta \( x[n] \) periódica con periodo \( N \):

$$
P = \frac{1}{N} \sum_{n=0}^{N-1} |x[n]|^2
$$

# 1) Desarrollo y análisis

<p align="center">
<p style="text-indent: 40px;">Para comenzar mi primer intento de tp1 desarrollé funciones específicas para poder hacer las operaciones requeridas sobre la función seno base, como por ejemplo:</p></p>

<p align="center">
<p style="text-indent: 40px;">Luego de un par de clases, descubrí que ya existen funciones predeterminadas de numpy o scipy.signal que ahorran líneas de código y obtienen resultados más precisos. Además, con ayuda de mis compañeros descubrí un par de tips que lograron que mis gráficos se vean mejor, como generar una envolvente con rayas punteadas alrededor de la función modulada. En los gráficos se jugó con los límites de x e y para poder mostrar de manera óptima el resultado.</p>
<p style="text-indent: 40px;">En cuanto a las operaciones desconocidas, la función np.where evalúa si la condición t_pulso < pulso es verdadera, si lo es coloca un 1. Pero si es falsa, coloca un 0. De este modo se genera un pulso rectangular que vale 1 desde t = 0 hasta t = 10 ms, en este caso, y después vale 0.</p>
<p style="text-indent: 40px;">Adicionalmente, para modular la función base seno investigué qué es el parámetro m y qué es la frecuencia moduladora. El parámetro m es el índice de modulación en amplitud, es decir, controla la amplitud de la portadora (la señal de la función base) en función de la señal moduladora. Tener en cuenta que m va de 0 a 1. Y, por otro lado, la frecuencia moduladora (frecmoduladora) es la frecuencia de la señal que modula en amplitud a la portadora. En el gráfico se puede observar que la portadora queda envuelta por una sinusoidal de 1 kHz.</p>
<p style="text-indent: 40px;">En el gráfico se ve cómo la portadora queda envuelta por una envolvente sinusoidal de 1 kHz. Una vez obtenida la función modulada (x3) se le suma 1 a la moduladora para asegurarse de que la envolvente nunca cambie de signo. En general, esta parte del trabajo utilizó mucho del ts0, por ende, no se llegó a niveles altos de dificultad.</p></p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square

#Parametros
frec = 2000 #frecuencia en Hz de la señal principal
frec_muestreo = 50000  #frecuencia de muestreo [Hz]
Ts = 1/frec_muestreo   #tiempo entre muestras

# Vectores de tiempo usando arange
t_seno = np.arange(0, 0.005, Ts)       
t_cuadrada = np.arange(0, 0.005, Ts)   
t_pulso = np.arange(0, 0.02, Ts)        

#Una señal sinusoidal de 2KHz.
x1 = np.sin(2 * np.pi * frec * t_seno)

#Sus graficos
plt.plot(t_seno, x1, color="hotpink")
plt.title('Funcion seno base')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.xlim(0, 0.005)
plt.grid(True)
plt.show()

#Misma señal amplificada y desfazada en π/2
x2 = 2 * np.sin(2 * np.pi * frec * t_seno + np.pi/2)

#Sus graficos
plt.plot(t_seno, x2, color="hotpink")
plt.title('Señal amplificada y desfasada π/2')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.xlim(0, 0.005)  
plt.grid(True)
plt.show()

#Misma señal pero modulada en amplitud por otra de la mitad de frecuencia
frecmoduladora = 1000  # señal moduladora
m = 0.6  
x3 = (1 + m * np.sin(2 * np.pi * frecmoduladora * t_seno)) * x1

#Sus graficos
plt.plot(t_seno, x3, color="hotpink")
plt.title('Señal de 2 kHz modulada en amplitud por 1 kHz')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.xlim(0, 0.005)  
plt.grid(True)

#Dibujo la envolvente
envolvente_superior = 1 + m * np.sin(2 * np.pi * frecmoduladora * t_seno)
envolvente_inferior = -(1 + m * np.sin(2 * np.pi * frecmoduladora * t_seno))
plt.plot(t_seno, envolvente_superior, 'g--')
plt.plot(t_seno, envolvente_inferior, 'g--')
plt.show()

#Señal anterior recortada al 75% de su amplitud
A = np.max(np.abs(x1))   #valor max de amp o.g
Aclip = 0.75 * A        #le saco el 75%
x1_clippeada = np.clip(x1, -Aclip, Aclip)

#Sus graficos
plt.plot(t_seno, x1_clippeada, color="hotpink")
plt.title('Señal anterior recortada al 75 % de su potencia')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.xlim(0, 0.005)  
plt.grid(True)
plt.show()

#Señal cuadrada de 4 kHz
frec_cuadrada=4000
amp = 1    
x_cuadrada = amp * square(2 * np.pi * frec_cuadrada * t_cuadrada)

#Sus graficos
plt.plot(t_cuadrada, x_cuadrada, color="hotpink")
plt.title('Señal cuadrada de 4 kHz')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.xlim(0, 0.00125)  
plt.grid(True)
plt.show()

pulso = 0.01
x_pulso = np.where(t_pulso < pulso, 1, 0)  # vale 1 de 0 a 10 ms, 0 después

#Sus graficos
plt.plot(t_pulso, x_pulso, color="hotpink")
plt.title('Pulso rectangular de 10 ms')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.ylim(-0.1, 1.1)  # Para ver mejor los bordes
plt.xlim(0, 0.02)    # Para ver la transición  
plt.grid(True)
plt.show()


#me hago un diccionario con todas mis funciones asi puedo hacer un for, me ahorro lineas de codigo
signals = {
    "Funcion base seno": x1,
    "Seno amplificado y desfasado": x2,
    "Modulada ": x3,
    "Clippeada": x1_clippeada,
    "Cuadrada": x_cuadrada,
    "Pulso rectangular": x_pulso
}

def calc_potencia(x): 
    return np.mean(x**2)

def calc_energia(x, Ts): 
    return np.sum(x**2) * Ts

#A señales periódicas le calculo la potencia y a las señales de duración finita la energía

print("Energía, potencia y número de muestras de cada señal (Nro_m):")
for name, sig in signals.items():
    if (name == "Pulso rectangular"): #!= devuelve un booleano en numpy
        Pot = calc_potencia(sig)
        print(f"{name}: su potencia es {Pot} y tiene {len(sig)} muestras.")
    else:
        E_pulso = calc_energia(x_pulso, Ts)
        print(f"{name}: su energia es {E_pulso} y tiene {len(sig)} muestras.")
```

# 2) 3) 4) Desarrollo y análisis: ortogonalidad, correlación e identidad trigonométrica

<p align="center">
<p style="text-indent: 40px;">En cuanto a la ortogonalidad, para verificarla fue necesario recortar los vectores y dejar todas con la misma cantidad de muestras, así se puede comparar de forma correcta. Si esto no se hacía, np.dot no funcionaría, ya que al usarse para calcular el producto interno entre dos señales, solo puede operar con vectores de igual longitud. Los resultados fueron:</p></p>

- Funcion base seno producto con Seno amplificado y desfasado = -0.000000 --> Es ortogonal
- Funcion base seno producto con Modulada = 125.000000 --> No es ortogonal
- Funcion base seno producto con Clippeada = 106.789307 --> No es ortogonal
- Funcion base seno producto con Cuadrada = 9.390625 --> No es ortogonal

<p align="center">
<p style="text-indent: 40px;">El producto con el seno amplificado y desfasado tiene sentido que sea ortogonal, ya que al estar desfasado 90° respecto de la función base y al multiplicarse por un número entero de ciclos, la parte positiva y negativa se anulan entre sí. Luego en el producto de la función base con la modulada, no se anulan entre sí, es decir, no son ortogonales porque la moduladora introduce variaciones lentas, al hacerlo coincide parcialmente. Por otro lado, la "clipped", si bien su amplitud es recortada al 75%, la forma de onda sigue siendo parecida a la función base. Por ende se multiplican y suman los valores, no se cancelan y por este motivo no son ortogonales. Por último, la x_cuadrada contiene armónicos distintos a los de la función base. Pero al tener en cuenta un número entero de ciclos, la multiplicación casi se anula, dando como resultado una señal aproximadamente ortogonal.</p>

<p style="text-indent: 40px;">Con respecto a la correlación cruzada, que, según lo explicado en la introducción, permite medir el grado de similitud entre dos señales cuando una se va desplazando en el tiempo respecto de la otra. Este desplazamiento también es conocido como lag y se puede observar más claramente en los gráficos. Por un lado, un pico alto en la correlación indica una gran semejanza en ese desplazamiento, mientras que valores cercanos a cero representan poca similitud. El más parecido a la función base es la amplificada y desfasada, a medida que se grafican los otros esa similitud disminuye.</p>

<p style="text-indent: 40px;">Finalmente, se comprobó la identidad trigonométrica que relaciona un producto de senos con sumas y diferencias de cosenos, como se mencionó en la introducción. El código verificó la igualdad entre ambos lados de la expresión y el resultado fue consistente.</p></p>

```python
import numpy as np
import matplotlib.pyplot as plt
señales = {
    "Seno amplificado y desfasado": x2,
    "Modulada": x3,
    "Clippeada": x1_clippeada,
    "Cuadrada": x_cuadrada[:len(t_seno)]  

}

print("Ortogonalidad entre x1 y las demás señales:")
for nombre, sig in señales.items():
    sig_recortada = sig[:len(x1)]
    producto = np.dot(x1, sig_recortada)
    
    if np.isclose(producto, 0, atol=1e-6):
        resultado = "Es ortogonal"
    else:
        resultado = "No es ortogonal"
    print("Funcion base seno producto con", nombre, "=", f"{producto:.6f}", "-->", resultado)

#autocorrelación de x1
autoco_x1 = np.correlate(x1, x1, mode='full')

# correlación de x1 con otras señales
correlaciones = {}
for nombre, sig in señales.items():
    sig_recortada = sig[:len(x1)]  # recortamos si es necesario
    correlacion = np.correlate(x1, sig_recortada, mode='full')
    correlaciones[nombre] = correlacion

lags = np.arange(-len(x1)+1, len(x1))  

# autocorrelación
plt.plot(lags, autoco_x1, color="hotpink")
plt.title("Autocorrelación de x1")
plt.xlabel("Lag")
plt.ylabel("Valor")
plt.grid(True)
plt.show()

# correlación con otras señales
for nombre, correlacion in correlaciones.items():
    plt.plot(lags, correlacion, label=nombre)
plt.title("Correlación de funcion base seno con otras señales")
plt.xlabel("Lag")
plt.ylabel("Valor")
plt.grid(True)
plt.legend()
plt.show()

frecuencia = 200  
w = 2 * np.pi * frecuencia  # Frecuencia angular ω
duracion = 0.01  # 10 ms de señal
frec_muestreo = 44100  # Frecuencia de muestreo
cant_muestras = int(duracion * frec_muestreo)

# Crear vector de tiempo
t = (np.arange(cant_muestras) / frec_muestreo).reshape(-1, 1)

alpha = w * t
beta = 2 * alpha 

# Calcular ambos lados de la identidad
ecuacion1= 2 * np.sin(alpha) * np.sin(beta)
ecuacion2 = np.cos(alpha - beta) - np.cos(alpha + beta)

# Verificar la igualdad (con tolerancia para errores numéricos)
if np.allclose(ecuacion1, ecuacion2, atol=1e-10):
    print("La identidad trigonométrica se cumple!")
else:
    print("La identidad no se cumple :(")
```

<p align="center">
<p style="text-indent: 40px;">Con respecto al bonus, no me parecio complejo el codigo en si, pero tuve que utilizar funciones o operaciones que no fueron vistas con anterioridad. Por un lado, fue necesario convertir la señal de estereo a mono, estereo se refiere a dos señales de audio, una en el canal izquierdo y otra en el derecho. Esto permite un sonido espacial, con profundidad y direccion. Y el mono, es una sola señal por un solo canal, es decir, que todos los altavoces reproducen el mismo sonido. La conversion se debe a que al ser por un solo canal facilita el analisis y permite evitar duplicar la energia al sumar ambos canales. Se ejecuto usando datos_audio.shape para devolver los datos de audio con dimensiones o "forma" de un array. Luego, datos_audio.mean(axis=1) calcula el promedio fila por fila, es otras palabras, entre los canales izquierdo y derecho.</p>
<p style="text-indent: 40px;">Finalmente, como ultima consideración, se normalizo entre -1 y 1, porque se complica trabajar con valores tan grandes. Esto se llevo a cabo transformando a flotantes de 32 bits, ya que como se quiere dividir se esperan decimales. Despues np.max(np.abs(datos_audio) te da el valor absoluto mas grande. Al dividir cada muestra por ese máximo absoluto, todos los valores quedan dentro del rango -1 a 1. No esta de mas aclarar que la importacion y manejo de archivos, graficar, calcular energia y el vector tiempo fueron desarrollados previamente.</p></p>

```python
import scipy.io.wavfile as wavfile

def analizar_audio(archivo_wav):
    
    # Leer el archivo WAV
    frecuencia_muestreo, datos_audio = wavfile.read(archivo_wav)
    
    # Convertir a mono si es estéreo
    if len(datos_audio.shape) > 1:
        datos_audio = datos_audio.mean(axis=1)
        print("Convertido a mono")
    
    # Normalizar entre -1 y 1 
    datos_normalizados = datos_audio.astype(np.float32) / np.max(np.abs(datos_audio))
    # si no normalizo son valores muy grandes
    
    # Calcular energía
    energia = np.sum(datos_normalizados**2)
    print(f"Energía total: {energia:.4f}")
    
    # Crear vector de tiempo
    tiempo = np.arange(len(datos_normalizados)) / frecuencia_muestreo
    
    # Graficar
    plt.plot(tiempo, datos_normalizados, color="hotpink")
    plt.title('Bonus: señal de audio de freesound.org')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud normalizada')
    plt.grid(True)
    plt.show()
    
    return datos_normalizados, frecuencia_muestreo, energia

archivo = "soundwav.wav" 
datos, fs, energia = analizar_audio(archivo)
```
