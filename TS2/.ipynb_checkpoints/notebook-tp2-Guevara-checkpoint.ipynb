# Introducción

En este trabajo se requiere que, a través de diferentes ecuaciones en diferencias, se discutan, analicen y grafiquen las señales de entrada, salida y respuesta al impulso. Antes de comenzar, se dará una breve introducción a estos conceptos con el propósito de brindar un marco conceptual claro y ordenado.

Un sistema se caracteriza por tener una **entrada *x[n]***, una **salida *y[n]*** y una **transformacion *T{}***, que se aplica a la entrada para producir la salida, es decir, *y[n]=T{x[n]}*. En el caso particular de que la entrada sea un impulso, *x[n]=δ[n]*, la salida recibe un nombre especial, "respuesta al impulso".y se asigna a la variable *h[n]*, esto sería: *h[n]≜T{δ[n]}*. *δ[n]* es una señal que vale 1 en el instante 0 y 0 en todos los demás instantes. Para visualizar esto, observar el esquema *(Esquema 1)* extraído del Holton*. 

Los **sistemas LTI**, lineales e invariantes en el tiempo, se caracterizan por responder de manera predecible y constante ante cualquier entrada. En esta practica se proporciono la ecuación 1. Hay dos tipos de sistemas LTI, **Respuesta al impulso finita** donde el sistema tiene respuestas al impulso finitas en duracion de tiempo *(Ecuación 2)* y **Respuesta al impulso infinita** donde el sistema tiene respuestas al impulso infinitas en duracion de tiempo *(Ecuación 3)*.
    
Una operación matemática utilizada en este trabajo es la convolución, que es una operación que se realiza a dos secuencias y se representa con un asterisco "*". A continuación se denotan la ecuación (Ecuación 4) y su esquema (Esquema 2). Esto es práctico porque, si conocés la respuesta al impulso, h[n], podés calcular la salida para cualquier entrada aplicando una única operación, la convolución.
    
Para poder clasificar y cuantificar señales periódicas o no periódicas, se utilizó la **energía total** *(Ecuación 5)* y la **potencia promedio** *(Ecuación 6)*. Por un lado, las señales no periódicas se describen por su energía total, que representa la “fuerza acumulada” a lo largo del tiempo. Mientras que las señales periódicas se caracterizan por su potencia promedio, que indica la energía transmitida por unidad de tiempo.

#### **Ecuación 1: ecuación en diferencias que modela un sistema LTI**
$$
y[n]=3\cdot 10^{-2}x[n]+5\cdot 10^{-2}x[n-1]+3\cdot 10^{-2}x[n-2]+1.5\,y[n-1]-0.5\,y[n-2]
$$

#### **Ecuación 2: Sistema FIR (Finite Impulse Response)**
$$
y[n] = \sum_{k=0}^{M} b_k \, x[n-k]
$$

#### **Ecuación 3: Sistema IIR (Infinite Impulse Response)**
$$
y[n] = \sum_{k=0}^{M} b_k \, x[n-k] \;-\; \sum_{j=1}^{N} a_j \, y[n-j]
$$

#### **Ecuación 5: Energía total de una señal discreta**
Para una señal discreta \( x[n] \) de duración finita o no periódica:

$$
E = \sum_{n=-\infty}^{\infty} |x[n]|^2
$$

#### **Ecuación 6: Potencia promedio de una señal periódica**
Para una señal discreta \( x[n] \) periódica con periodo \( N \):

$$
P = \frac{1}{N} \sum_{n=0}^{N-1} |x[n]|^2
$$

Comienzo el código copiando las señales de entrada que generó en el TS1. Luego se arma una función de la ecuación en diferencias que modela un sistema LTI, considerando que son causales. Causal se refiere a que la salida en un instante n depende solamente de la entrada en ese mismo instante n o en instantes pasados *(n−k)*, pero nunca de valores futuros *(n+k con k>0)*, esto se logra a través de un if que respete lo de causal. Adicionalmente, tanto para el primer y segundo bulletpoint se indicó la frecuencia de muestreo, el tiempo de simulación y la potencia o energía de la señal de salida. El tiempo de simulación siendo la diferencia entre *t[0]* = instante inicial (≈ 0) y *t[-1]* = último instante simulado. La parte compleja del trabajo comienza al realizar la operación de convolución. Se puede convolucionar de dos maneras:

a) Generando un delta, lo filtrás y obtenés *h[n]*. Este luego se convoluciona con las señales, *x[n]*. Finalmente mostrás que la salida coincide con la obtenida directamente del sistema.

b) De manera más teórica, sacando un H(Z) (que se puede hacer dividiendo coeficientes de x con los coeficientes de y), evaluando en el círculo unitario $$ z = e^{j \omega} $$ y antitransformando para hallar *h[n]*.

La forma utilizada en este trabajo es el a), debido a su simplicidad.

En el punto dos se hicieron dos funciones separadas para el sistema FIR y el sistema IIR. Luego con la función delta se calculó la respuesta al impulso. # 1) Codigo y gráficos

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square

# ================================================================
# PUNTO 1
# ================================================================

# las señales de entrada que generó en el TS1
frec = 2000  # Hz
frec_muestreo = 50000  # Hz
Ts = 1/frec_muestreo

t_seno = np.arange(0, 0.005, Ts)       
t_cuadrada = np.arange(0, 0.005, Ts)   
t_pulso = np.arange(0, 0.02, Ts)        

# Señales de entrada
x1 = np.sin(2 * np.pi * frec * t_seno)
x2 = 2 * np.sin(2 * np.pi * frec * t_seno + np.pi/2)

frecmoduladora = 1000  
m = 0.6
x3 = (1 + m * np.sin(2 * np.pi * frecmoduladora * t_seno)) * x1

A = np.max(np.abs(x1))   
Aclip = 0.75 * A        
x1_clippeada = np.clip(x1, -Aclip, Aclip)

frec_cuadrada = 4000
amp = 1    
x_cuadrada = amp * square(2 * np.pi * frec_cuadrada * t_cuadrada)

pulso = 0.01
x_pulso = np.where(t_pulso < pulso, 1, 0)

# ecuación en diferencias que modela un sistema LTI, considerando que son causales
def sistema_LTI(x):
    N = len(x)
    y = np.zeros(N)
    for n in range(N):
        # entrada
        xn   = 0.03*x[n]
        xn1  = 0.05*x[n-1] if n-1 >= 0 else 0
        xn2  = 0.03*x[n-2] if n-2 >= 0 else 0
        # salida
        yn1  = 1.5*y[n-1] if n-1 >= 0 else 0
        yn2  = -0.5*y[n-2] if n-2 >= 0 else 0
        y[n] = xn + xn1 + xn2 + yn1 + yn2
    return y

# calculo salidas
y1 = sistema_LTI(x1)
y2 = sistema_LTI(x2)
y3 = sistema_LTI(x3)
y1_clip = sistema_LTI(x1_clippeada)
y_sq = sistema_LTI(x_cuadrada)
y_pulso = sistema_LTI(x_pulso)

# grafico
fig, axs = plt.subplots(6, 2, figsize=(12, 12))

signals = [
    ("x1 seno 2kHz", t_seno, x1, y1),
    ("x2 seno 2kHz amplif/fase", t_seno, x2, y2),
    ("x3 AM", t_seno, x3, y3),
    ("x1 clippeada", t_seno, x1_clippeada, y1_clip),
    ("x_cuadrada 4kHz", t_cuadrada, x_cuadrada, y_sq),
    ("x_pulso 10ms", t_pulso, x_pulso, y_pulso),
]

for i, (name, t, x, y) in enumerate(signals):
    axs[i,0].plot(t, x)
    axs[i,0].set_title(f"Entrada {name}")
    axs[i,0].grid(True)
    axs[i,1].plot(t, y)
    axs[i,1].set_title(f"Salida {name}")
    axs[i,1].grid(True)

plt.tight_layout()
plt.show()

'''
me piden "En cada caso indique la frecuencia de muestreo, 
el tiempo de simulación y la potencia o energía de la señal de salida."
'''
def calc_potencia(x): 
    return np.mean(x**2)

def calc_energia(x, Ts): 
    return np.sum(x**2) * Ts

def printear_datos():
    print("Frecuencia de muestreo, tiempo de simulación y potencia/energía de la salida:")
    for nombre, t, sig, y in signals:
        tiempo_sim = t[-1] - t[0]
        if "pulso" in nombre:  # energía si es de duración finita
            E = calc_energia(y, Ts)
            print(f"{nombre}: fs = {frec_muestreo} Hz, Tsim = {tiempo_sim:.5f} s, Energía = {E:.5f}")
        else:                  # potencia si es periódica
            P = calc_potencia(y)
            print(f"{nombre}: fs = {frec_muestreo} Hz, Tsim = {tiempo_sim:.5f} s, Potencia = {P:.5f}")
#t[0] = instante inicial (≈ 0), t[-1] = último instante simulado.
#el intervalo total de tiempo de la simulación es la diferencia

print("Datos primer bulletpoint:\n")
printear_datos()
print("\n")

# Hallar la respuesta al impulso 
N = 50
delta = np.zeros(N)
delta[0] = 1
h = sistema_LTI(delta)

#grafico rta al impulso
n = np.arange(N)  
plt.figure(figsize=(10,4))
plt.stem(n, h)
plt.title("Respuesta al impulso h[n]")
plt.xlabel("n")
plt.ylabel("h[n]")
plt.grid(True)
plt.show()

# convolucion manual
def convolucion(u):
    y = np.zeros(len(u))
    for n in range(len(u)):
        suma = 0
        for k in range(len(h)):
            if n-k >= 0:
                suma += h[k] * u[n-k]
        y[n] = suma
    return y

#convoluciono tutti y las grafioc
for nombre, t, sig, y in signals:
    # Salidas
    y1_directa = sistema_LTI(sig)
    y1_conv = convolucion(sig)[:len(sig)]   # recorto para que tenga la misma longitud
    
    # Gráfico comparativo
    plt.figure(figsize=(10,5))
    plt.plot(t, y1_directa, label="Salida directa (sistema)", color="b")
    plt.plot(t, y1_conv, "--", label="Salida por convolución", color="r")
    plt.title(f"Comparación de salidas para {nombre}")
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud")
    plt.legend()
    plt.grid(True)
    plt.show()
    
print("Datos segundo bulletpoint:\n")
printear_datos()
```

# 1) Análisis gráficos
Se graficaron la entrada y la salida, para poder compararlos con mayor eficiencia y realizar un análisis. Primeramente, para x1 y x2 el sistema se comportó como un filtro lineal. Se ve un cambio leve, desfase y cambio de amplitud.

En cuanto a x3, la señal modulada, el sistema afecta más a ciertas frecuencias que a otras actuando como un filtro selector. Concretamente lo que sucede es que x3 es una portadora de 2 kHz con una moduladora de 1 kHz, esto sería un espectro con una componente en 2 kHz y dos laterales en 1 kHz y 3 kHz. El sistema no es plano, no amplifica por igual 1, 2 y 3 kHz. Por ende, la envolvente de la señal modulada se distorsiona porque las laterales no tienen la misma ganancia que la portadora.

La señal recortada a simple vista parece que dejó de estar recortada. Esto se debe a que como la señal recortada contiene muchísimos armónicos impares más que componentes fundamentales, el sistema parece atenuar o modificar las altas frecuencias más que a la fundamental, suavizando la señal recortada.

La onda cuadrada por otro lado está hecha de una senoidal fundamental (4 Hz) más infinitos armónicos impares. Una de las preguntas que surgieron fue por qué se hace senoidal creciente, algo que trataré de explicar a continuación. El sistema amplifica algunas frecuencias y otras las atenúa o cancela, pero normalmente los armónicos más altos de la cuadrada son los atenuados, por eso sobrevive la componente fundamental que es senoidal. Luego que sea creciente se debe a que el sistema tiene polo con ganancia mayor a 1, es decir que la salida no sea estable y crezca con el tiempo.

Finalmente, el pulso contiene muchas frecuencias y el sistema responde más a unas frecuencias que a otras, deformando así el pulso. Además, el pulso excita la respuesta natural del sistema (los polos), provocando que al inicio se pueda invertir o oscilar y después estabilizarse en una cola más larga.

En los gráficos de convolución, en todos se respeta que comparten una parte y después no, esta pregunta se trató de explicar a continuación. Al principio (las primeras muestras), la convolución manual y la ecuación en diferencias usan casi los mismos valores y la diferencia es mínima, por eso coinciden esas primeras “montañitas”. Se recortó la respuesta al impulso a solo 50 muestras (N=50), pero el sistema real sigue teniendo respuesta más allá de esas 50. Es por esto que h está incompleta, truncada. Si se convoluciona con un h[n] truncado al principio se parece mucho, ya que los primeros valores son los más fuertes. Luego empieza a diferir porque al sistema real le sigue “resonando” memoria de salidas pasadas, y eso no aparece en la convolución truncada.

# 2) Codigo y gráficos

```python
import numpy as np
import matplotlib.pyplot as plt
# ================================================================
# PUNTO 2
# ================================================================

# rta al impulso y la salida correspondiente a una señal de entrada senoidal 
#en los sistemas definidos mediante las siguientes ecuaciones en diferencias:

def sistema1(x):
    N = len(x)
    y = np.zeros(N)
    for n in range(N):
        xn   = x[n]
        xn10 = 3*x[n-10] if n-10 >= 0 else 0
        y[n] = xn + xn10
    return y

def sistema2(x):
    N = len(x)
    y = np.zeros(N)
    for n in range(N):
        xn = x[n]
        yn10 = 3*y[n-10] if n-10 >= 0 else 0
        y[n] = xn + yn10
    return y

# TIEMPOS Y SEÑALES PARA CADA SISTEMA

# SIST 1 (FIR) con menos muestras para ver la senoidal claramente
t1 = np.linspace(0, 0.003, 60)  # menos tiempo
x1 = np.sin(2*np.pi*1000*t1)    
# SIST 2 (IIR) con más muestras para ver el crecimiento exponencial
t2 = np.linspace(0, 0.05, 120)  # mas tiempo
x2 = np.sin(2*np.pi*1000*t2)

# RTA AL IMPULSO
N_impulso = 50
delta = np.zeros(N_impulso)
delta[0] = 1
h1 = sistema1(delta)
h2 = sistema2(delta)

# SALIDAS
y1 = sistema1(x1)
y2 = sistema2(x2)

plt.figure(figsize=(12, 8))

#GRAFICO 1: Respuesta al impulso - Sistema 1
plt.subplot(2, 2, 1)
n = np.arange(len(h1))
plt.stem(n, h1)
plt.title("Respuesta al impulso - Sistema 1 (FIR)")
plt.xlabel("Muestra [n]")
plt.ylabel("Amplitud")
plt.grid(True)

#GRAFICO 2: Salida senoidal - Sistema 1 CORREGIDO
plt.subplot(2, 2, 2)
plt.plot(t1*1000, y1)  # Convertir a milisegundos
plt.title("Salida para senoidal - Sistema 1")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.grid(True)

#GRAFICO 3: Respuesta al impulso - Sistema 2
plt.subplot(2, 2, 3)
n = np.arange(len(h2))
plt.stem(n, h2)
plt.title("Respuesta al impulso - Sistema 2 (IIR)")
plt.xlabel("Muestra [n]")
plt.ylabel("Amplitud")
plt.grid(True)

# GRAFICO 4: Salida senoidal - Sistema 2 CORREGIDO
plt.subplot(2, 2, 4)
plt.plot(t2*1000, y2)  #convertir a milisegundos
plt.xlim(0,48)
plt.title("Salida para senoidal - Sistema 2")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# 2) Análisis gráficos
El sistema 1 clasifica dentro de FIR, finite impulse response, ya que depende solo de entradas pasadas. En el gráfico de respuesta al impulso se puede observar un 1 en n=0 y un 3 en n=10. Esto muestra que el sistema retrasa 10 muestras y las amplifica por 3, el impulso es “especial” porque solo vale 1 en n=0 y es 0 en todos los demás instantes, por eso tiene "dos palitos". Para el gráfico de la salida al principio hay "dos montañitas", una en 1 y otra en 2, esto se debe a que al principio la salida ve a la señal original, todavía no hay copia retrasada (no hay valores para n-10). Cuando ya conviven la señal original y la copia retrasada todo el tiempo, la forma de onda se estabiliza en algo senoidal modificado.

El sistema 2 clasifica dentro de IIR, infinite impulse response, ya que depende de salidas pasadas. Según el gráfico de la respuesta al impulso y un poco de zoom, se puede observar que para n=0 vale 1, para n=10 aproximadamente 3, luego para n=20 es 9 (3 elevado al cuadrado), el siguiente para n=30 es 27 (3 elevado al cubo) y así con los otros n. Esto se ve en el gráfico de la salida, que comienza oscilando a valores chicos y después se dispara rápido en valores grandes. El motivo de su inestabilidad es que tiene polos fuera del círculo unitario, esto explica por qué en la salida para la señal senoidal la amplitud se "desborda numéricamente", alcanzando valores del orden de 1e11, es decir, la realimentación positiva amplifica exponencialmente cualquier entrada.

En base a la corrección, a continuación se explaya detenidamente el motivo de los tiempos de simulación elegidos y el criterio utilizado.
Para el primer sistema se seleccionó un tiempo de 3 ms (60 muestras), con una respuesta al impulso de longitud finita, para visualizar claramente 3 períodos completos de la señal senoidal de 1 kHz, permitiendo observar el comportamiento estable del sistema. Con un tiempo mayor no se podría apreciar bien la forma de la onda, ya que se visualizaría toda comprimida y para un tiempo menor, no hubiera mostrado suficientes períodos para confirmar la estabilidad. Para el segundo sistema se optó por un tiempo más extenso de 50 ms (120 muestras), que permite observar claramente la transición desde valores pequeños iniciales hasta el crecimiento exponencial característico de la inestabilidad, debido a que su amplitud se triplica cada 10 muestras. Este tiempo muestra de manera evidente cómo el sistema se va de escala progresivamente.

Se tuvo en cuenta:
- No usar el mismo tiempo para todos los sistemas: Cada sistema tiene su propia longitud óptima.
- Se basó en la respuesta al impulso.
- Se consideró el orden del sistema.
- Memoria necesaria: Sistema 1 necesita 10 muestras de memoria, Sistema 2 necesita memoria infinita teóricamente.
